5. Write a ``ChargedParticle`` class that inherits from the ``Particle`` class, but also has an attribute for the charge of the particle.


```python
import numpy as np
class Particle(object):
    def __init__(self,m,x,y,z,vx,vy,vz):
        self.m = m
        self.x = x
        self.y = y
        self.z = z
        self.vx = vx
        self.vy = vy
        self.vz = vz
        
    def K_E(self):
        return (1/2)*self.m*(self.vx*2+self.vy*2+self.vz*2)
        
    def find_separation(self, other):
        return np.sqrt((self.x - other.x)**2 + (self.y - other.y)**2 + (self.z - other.z)**2)
        
    def pos_update(self,dt):
        self.x = self.x + self.vx*dt
        self.y = self.y * self.vy*dt
        self.z = self.z * self.vz*dt
        return(self.x,self.y,self.z)


class ChargedParticle(Particle):
    
    def __init__(self, mass, x, y, z, vx, vy, vz, charge):
    
        super().__init__(mass, x, y, z, vx, vy, vz)
        # Add a new attribute for electric charge
        self.charge = charge


p1 = Particle(2,0,0,0,5,0,0)
p2 = Particle(10,0,4,0,0,11,2)
p3 = ChargedParticle(10,0,4,0,0,11,2,-2)


    
print(p1.K_E())
print(p1.find_separation(p2))
print(p1.pos_update(3))
```
